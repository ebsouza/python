**6. OrdenaÃ§Ã£o do dataframe tendo a coluna 'math score' como referÃªncia**

```python
"""
Para ordenar em ordem crescente, basta declarar "df.sort_values('math score')"

Este comando nÃ£o modifica o dataframe referenciado por "df"
"""
df.sort_values('math score', ascending=False)
```

**7. Criando um novo score baseado em 'math score', 'reading score' e 'writing score'.**

**[FÃ³rmula de 'new score']**

**new score = (30% x 'math score') + (50% x 'reading score') + (20% x 'writing score')**

```python
df['new score'] = 0.3*df['math score'] + 0.5*df['reading score'] + 0.2*df['writing score']

df
```

**8. Criando subdataframe: estudantes com 'new score' maior que 80**

```python
"""
Apenas 200 estudantes possuem 'new score' maior que 80
"""

sub_dataframe = df[df['new score'] > 80]

sub_dataframe
```
